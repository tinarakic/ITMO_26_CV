```python
import os
import shutil
import random

# Path to Kaggle dataset
dataset_dir = '/kaggle/input/datasets/kport354041/honeybee-positions'

# Input folders
frames_dir = os.path.join(dataset_dir, 'frames')
labels_dir = os.path.join(dataset_dir, 'positions')

# Output folders
output_dir = '/kaggle/working/honeybee-sample-dataset'
output_frames_dir = os.path.join(output_dir, 'frames')
output_labels_dir = os.path.join(output_dir, 'positions')

# Number of random samples
num_samples = 100

# Remove old output directory if it exists
if os.path.exists(output_dir):
    shutil.rmtree(output_dir)

# Create fresh output directories
os.makedirs(output_frames_dir)
os.makedirs(output_labels_dir)

# Get all frame files
frame_files = [
    f for f in os.listdir(frames_dir)
    if os.path.isfile(os.path.join(frames_dir, f))
]

# Randomly select 100 frames
selected_frames = random.sample(
    frame_files,
    min(num_samples, len(frame_files))
)

# Copy selected frames and corresponding labels
for frame_file in selected_frames:
    
    # Copy frame
    src_frame = os.path.join(frames_dir, frame_file)
    dst_frame = os.path.join(output_frames_dir, frame_file)
    shutil.copy(src_frame, dst_frame)

    # Find corresponding label file
    base_name = os.path.splitext(frame_file)[0]
    label_file = base_name + '.txt'

    src_label = os.path.join(labels_dir, label_file)

    # Copy label if it exists
    if os.path.exists(src_label):
        dst_label = os.path.join(output_labels_dir, label_file)
        shutil.copy(src_label, dst_label)

print(f"Copied {len(selected_frames)} random samples.")
print("Output saved to:", output_dir)

# Create ZIP archive
zip_path = '/kaggle/working/honeybee-sample-dataset.zip'

shutil.make_archive(
    base_name=zip_path.replace('.zip', ''),
    format='zip',
    root_dir=output_dir
)

print("ZIP archive created at:")
print(zip_path)
```
